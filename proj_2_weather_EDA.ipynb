```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
# Create dataframe
df = pd.read_csv("Weather_Data.csv")
df.head()
```

#### EDA and Data Cleaning

```python
# Displaying the key characteristics of the dataset (e.g., number of records, features, data types)
df.info()
```

```python
# Identify and handle any missing or null values.
df.isnull().sum()
```

```python
# Check for and address any duplicate records
df.duplicated().sum()
```

```python
# Statistical summary of the dataset (mean, median, standard deviation, etc.) for numerical features.
df.describe()
```

```python
# Identifying and describing any significant outliers in the data
sb.boxplot(data=df)
plt.xticks(rotation=45)
plt.xlabel("Weather Features")
plt.ylabel("Measurement")
```

The boxplot above shows outliers of about 50km visibility and 85 km/h wind speed. This may be due to errors in measurement or represent extreme cases.  

#### Data Visualisation
Distribution of key weather parameters (e.g., temperature, humidity, wind speed).

```python
df.columns
```

```python
# Temperature distribution
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

sb.histplot(df["Temp_C"], ax=axes[0], kde=True)
axes[0].set_title('Histplot')

sb.kdeplot(df["Temp_C"], ax=axes[1], fill=True)
axes[1].set_title('kdeplot')

# Adjust layout to prevent overlap
plt.tight_layout()
```

```python
# Time series graphs to visualize trends over time.......... Highlight any notable patterns or seasonal variations.

# Set the datetime column as the index
df.set_index('Date_Time', inplace=True)

# Plot Temperature over Time
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Temp_C'], label='Temperature (°C)')
plt.xlabel('Date and Time')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_num = df.select_dtypes(include=[float, int])
corr_matrix = df_num.corr()
corr_matrix
```

```python
# Correlation matrices and heatmaps to identify relationships between different weather parameters
df_num = df.select_dtypes(include=[int,float])
sb.heatmap(data=df_num.corr(), annot=True, cmap='coolwarm')
```

The heat map above shows there is a high corelation between Temperature and Dew point temperature. Following the principle of muticollinearity, one of these independent variables will have to be dropped to prevent redundancy.
